```python
# Examining inconsistencies in sale_price column

df[df.LoanAmount.isnull()]
```

Find out the outliers from the dataset according to the columns and fill them with the Median.¶

```python
# Using fillna function to fill the null values present in the numeric columns with median of that particular column. 

df.LoanAmount  = df.LoanAmount.fillna(df.LoanAmount.median())
df.Loan_Amount_Term = df.Loan_Amount_Term.fillna(df.Loan_Amount_Term.median())
df.Credit_History=df.Credit_History.fillna(df.Credit_History.median())
```

```python
# Using fillna function to fill the null values present in the object columns with Unknown in that particular column.

df.Gender = df.Gender.fillna('Unknown')
df.Self_Employed = df.Self_Employed.fillna('Unknown')
df.Dependents  = df.Dependents.fillna('Unknown')
```

```python
# check null values after filling the columns.

df.isnull().sum()
```

-  The entire dataset is Now  void of all the null values.
